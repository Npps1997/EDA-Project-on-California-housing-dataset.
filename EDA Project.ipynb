

California Housing dataset Data Analysis Project
--------------------------
* The California Housing Dataset Analysis project aims to explore and understand various aspects of the California housing market. The dataset contains information about different housing districts in California and various features associated with each district (Block).

* Objective: Uncover insights and patterns within the California housing market through data analysis.

* Dataset: A comprehensive collection of housing attributes, encompassing various regions in California.

* This dataset was derived from the 1990 U.S. census, using one row per census block group. A block group is the smallest geographical unit for which the U.S. Census Bureau publishes sample data (a block group typically has a population of 600 to 3,000 people).

* A household is a group of people residing within a home.

**Data Set Characteristics:**

    :Number of Instances: 20640

    :Number of Attributes: 9 numeric, 1 categorical

    :Attribute Information:
        - Longitude           :block group longitude (A measure of how far west a house is; a higher value is farther west).
        - Latitude            :block group latitude (A measure of how far north a house is; a higher value is farther north).
        - housing_median_age  :median house age in block group (a lower number is a newer building).
        - total_rooms         :Total number of rooms within a block.
        - total_bedrooms      :Total number of bedrooms within a block.
        - population          :Total number of people residing within a block.
        - households          :Total number of households, a group of people residing within a home unit, for a block.
        - median_income       :median income in block group (measured in tens of thousands of US Dollars)
        - median_house_value  :median house value in block group (in US Dollars)
        - ocean_proximity     :Location of the house w.r.t ocean/sea



    :Attribute Classification: (In terms of QUANTITATIVE-discrete and continuous ; QUALITATIVE data - nominal and ordinal)

        -longitude (float)        :Continuous data (As they have decimal values and vary continuously within a specified range).

        -latitude (float)         :Continuous data (As they have decimal values and vary continuously within a specified range).

        -housing_median_age (int) :Continuous data (As they can take on any value within a range)

        -total_rooms (int)        :Discrete data (Data points are whole numbers)

        -total_bedrooms (int)     :Discrete data (Data points can be repeated number of times)

        -population (int)         :Discrete data (As data points are whole numbers

        -households (int)         :Discrete (They represent countable, distinct values)

        -median_income (float)    :Continuous data (As they can take on any value within a range)

        -median_house_value (int) :Continuous data (As they can take on any value within a range)

## 6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.

```python
#Creating a function cal_function to calculate median.

def cal_median(data):                                #Function definition by passing list data as argument
    sorted_data = sorted(data)                         #sorting list in ascending order and stored in variable 'sorted_data'
    n = len(sorted_data)                               #calculating lenght of sorted data and storing in variable 'n'

# Checking for even and odd lenght
    if n%2 == 1:                                       #checking if lenght is odd
        median_value = sorted_data[n//2]                 #storing the middle value of list in variable 'median_value'
    else:
        middle_left = sorted_data[(n//2)-1]              #if lenght is even, storing middle left value to 'middle_left'
        middle_right = sorted_data[n//2]
        median_value = (middle_left + middle_right)/2    #calculating average of two middle values

    return median_value                                #returning median
```

```python
#calling function by passing median_income column and printing the median value

median = cal_median(housing['median_income'])
print("Median value:", median)
```

## 10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python
#Creating a new column 'total_bedroom_size' for housing dataframe.

for i in housing.index:
    if housing.loc[i, 'total_bedrooms'] <= 10:
        housing.loc[i, 'total_bedroom_size'] = 'small'
    elif 11 <= housing.loc[i, 'total_bedrooms'] < 1000:
        housing.loc[i, 'total_bedroom_size'] = 'medium'
    else:
        housing.loc[i, 'total_bedroom_size'] = 'large'
```
